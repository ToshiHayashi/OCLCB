```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
from PIL import Image
import glob
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import jaccard_score,log_loss, mean_absolute_error,median_absolute_error, mean_squared_error
%matplotlib inline
from tqdm import tqdm
import pickle
def makefile(what,filename):
    with open(filename,"wb") as f3:
        pickle.dump(what,f3)

def readfile(filename):
    with open(filename,"rb") as f4:
        ans=pickle.load(f4)
    return ans
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
from sklearn.metrics import jaccard_score,log_loss, mean_absolute_error,median_absolute_error, mean_squared_error
from sklearn.metrics import roc_auc_score
from sklearn.naive_bayes import CategoricalNB
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import accuracy_score
from sklearn.metrics import balanced_accuracy_score
```

```python
v1=[1,3,6,7,8,9,11,13,17]
WS=[2,3,4,5,6,7,8,9,10]
cl=[2,3,5,10,15,20,25,50]
LS=[50,100,500,660,1000,2000,5000,10000]
```

```python

v1=[1,3,6,7,8,9,11,13,17]
WS=[2]
cl=[25]
LS=[660]


change=[60,120,180,240,300,420,480,540,600,660]
#change=[300,420]


for i in v1:
    name=str(i)
    for window_size in WS:
        s1=window_size
        for K in cl:
            for local_size in LS:
                errorname="UnseenScore/Error_"+name+"_K"+str(K)+"_"+str(local_size)+"_d"+str(s1)+".pkl"
                error=readfile(errorname)
                plt.title(errorname)
                plt.vlines(50000, error.min(),error.max(),color="green",linestyle = "dashed")
                for v in range(len(change)):
                    plt.vlines(change[v]*1000, error.min(),error.max(),color="red",linestyle = "dashed")
                plt.plot(error)
                plt.savefig("images/"+str(i)+".jpg")
                plt.figure()
```

```python
v1=[1]
WS=[2,3,4,5,6,7,8,9,10]
cl=[10]
LS=[5000]
n=0
fig=plt.figure(figsize=(20,15))
change=[60,120,180,240,300,420,480,540,600,660]

for i in v1:
    name=str(i)
    for window_size in WS:
        s1=window_size
        for K in cl:
            for local_size in LS:
                
                ax=fig.add_subplot(3,3,n+1)
                errorname="UnseenScore/Error_"+name+"_K"+str(K)+"_"+str(local_size)+"_d"+str(s1)+".pkl"
                error=readfile(errorname)
                index2=np.array([n for n in range(len(error))])/1000
                ax.vlines(50, error.min(),error.max(),color="green",linestyle = "dashed")
                for v in range(len(change)):
                    ax.vlines(change[v], error.min(),error.max(),color="red",linestyle = "dashed")
                errortitle="Window size = "+str(s1)
                ax.set_title(errortitle)
                ax.set_xlabel("time(seconds)")
                ax.set_ylabel("unseen score")
                ax.plot(index2,error)
                #plt.figure()
                n+=1
fig.savefig("images/fig4.jpg")
```
